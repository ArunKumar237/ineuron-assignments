Q-2. Write an SQL query to fetch duplicate records having matching data in some fields of a table.


```python
print('department column\n-------------------')
cursor.execute('SELECT department,count(department) FROM worker group by department having count(department) > 1')
cursor.fetchall()
# SELECT 
#     col, 
#     COUNT(col)
# FROM
#     table_name
# GROUP BY col
# HAVING COUNT(col) > 1;
```

Q-3. How to remove duplicate rows from Employees table.

```python
q = '''use test_data;
delete c1 from worker c1
join worker c2
where
	c1.Worker_id > c2.Worker_id and
    c1.First_name = c2.First_name'''
cursor.execute(q,multi=True)
cursor.fetchall()
```
